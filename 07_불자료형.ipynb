```python
# 불(bool) 자료형은 어떻게 사용할까?
a = True
b = False

print(a)
print(b)
print(type(a))
print(type(b))
```

```python
print(1 == 1)
print(2 > 1)
print(2 < 1)
print(2 >= 2)
print(2 <= 2)
print(2 != 2)
```

```python
# if 0:
#	 수행할 문장
print(bool("python"))
print(bool("")) ## 전혀 없는 거, 0만 아니면 True
print(bool(" "))
print(bool(10))
print(bool(0)) ## False
print(bool([])) ## False
print(bool([1, 2, 3]))
print(bool({"name":"ddd"}))
print(bool({})) ## False
print(bool()) ## False
```

```python
# 논리연산자

# and 연산자
print(True and True)
print(True and False)
print(False and True)
print(False and False)
print(True and True and False)
```

```python
# or 연산자
print(True or True)
print(True or False)
print(False or True)
print(False or False)
print(True or True or False)
```

```python
# not 연산자

print(not True)
```
