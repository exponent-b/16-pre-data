```python
'''
문제 7. 성적 관리 프로그램
학생들의 성적을 관리하는 프로그램을 작성하시오.

다음 학생들의 성적 정보를 딕셔너리로 저장하시오

학생1: 이름=김민수, 국어=85, 영어=90, 수학=78
학생2: 이름=이영희, 국어=92, 영어=88, 수학=95
학생3: 이름=박철수, 국어=78, 영어=85, 수학=82

요구사항:
1. 각 학생의 총점과 평균을 계산하여 딕셔너리에 추가
2. 각 학생의 정보를 보기 좋게 출력
출력 예시:

=== 학생 성적 관리 ===
이름: 김민수
국어: 85점, 영어: 90점, 수학: 78점
총점: 253점, 평균: 84.33점

이름: 이영희
국어: 92점, 영어: 88점, 수학: 95점
총점: 275점, 평균: 91.67점

이름: 박철수
국어: 78점, 영어: 85점, 수학: 82점
총점: 245점, 평균: 81.67점

'''
stus = {
	"학생1": {'이름': '김민수', '국어': 85, '영어': 90, '수학': 78},
	"학생2": {'이름': '이영희', '국어': 92, '영어': 88, '수학': 95},
	"학생3": {'이름': '박철수', '국어': 78, '영어': 85, '수학': 82}
	}


students = [
    {'이름': '김민수', '국어': 85, '영어': 90, '수학': 78},
    {'이름': '이영희', '국어': 92, '영어': 88, '수학': 95},
    {'이름': '박철수', '국어': 78, '영어': 85, '수학': 82}
]

for student in students:
    total = student['국어'] + student['영어'] + student['수학']
    avg = total / 3

    student['총점'] = total
    student['평균'] = avg

print("=== 학생 성적 관리 ===")

for student in students:
    print("이름:", student['이름'])
    print(f"국어: {student['국어']}점, 영어: {student['영어']}점, 수학: {student['수학']}점")
    print(f"총점: {student['총점']}점, 평균: {student['평균']:.2f}점\n")


```

```python
'''
문제 8. 단어 카운터
문자열에서 각 단어가 몇 번 등장하는지 세는 프로그램을 작성하시오.

text = "python is great python is fun python is powerful"

요구사항:
1. 문자열을 공백으로 분리하여 단어 리스트 만들기
2. 각 단어의 등장 횟수 세기
3. 결과를 출력

출력 예시:
python: 3번
is: 3번
great: 1번
fun: 1번
powerful: 1번
'''
text = "python is great python is fun python is powerful"

words = text.split()
# print(words)
count = {}

for word in words:
	if word in count:
		count[word] += 1
	else:
		count[word] = 1

print(count)
```

```python
'''
A반과 B반 학생 명단이 있습니다.

a_class = {'김철수', '이영희', '박민수', '최지영', '홍길동'}
b_class = {'박민수', '최지영', '정수민', '강호동', '이영희'}

다음 질문에 답하시오:

1. 두 반 모두에 있는 학생 (교집합)
2. 두 반을 합친 전체 학생 (합집합)
3. A반에만 있는 학생 (차집합)
4. B반에만 있는 학생 (차집합)
'''
a_class = {'김철수', '이영희', '박민수', '최지영', '홍길동'}
b_class = {'박민수', '최지영', '정수민', '강호동', '이영희'}

print(a_class & b_class)
print(a_class | b_class)
print(a_class - b_class)
print(b_class - a_class)
```

```python
'''
문제 12. 집합 조작하기

fruits = {'apple', 'banana', 'orange'}

다음 작업을 수행하는 코드를 작성하시오:

1. 'grape'를 추가하시오
2. 'melon', 'kiwi', 'mango'를 한 번에 추가하시오
3. 'banana'를 제거하시오
4. 집합에 'strawberry'가 있는지 확인하시오
5. 모든 과일을 제거하고 빈 집합으로 만드시오
'''
fruits = {'apple', 'banana', 'orange'}

fruits.add('grape')
fruits.update(['melon', 'kiwi', 'mango'])
fruits.remove('banana')
print('strawberry' in fruits)
fruits.clear()
print(fruits)
```
