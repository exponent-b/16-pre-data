```python
# 튜플은 어떻게 만들까? (tuple)
## 수정하거나 삭제가 불가

t1 = ()
t2 = (1, )
t3 = (1, 2, 3)
t4 = 1, 2, 3 # 괄호 제거 가능
t5 = ("a", "b", ("ab", "cd"))

print(type(t1))
print(type(t2))
print(type(t3))
print(type(t4))
print(type(t5))
```

```python
a = 1, 2
print(a) # 콤마를 써서 튜플이 됨
print(type(a))
```

```python
# 튜플의 요소값을 지우거나 변경하려고 하면 어떻게 될까? -> error

t1 = (1, 2, "a", "b")
del t1[0]
```

```python
# 튜플 요소값을 변경하려 할 때 -> error
t1 = (1, 2, "a", "b")
t1[0] = 2
print(t1)
```

```python
# 튜플 인덱싱

t1 = (1, 2, "a", "b")
print(t1[2])
print(t1[-1])
```

```python
# 튜플의 슬라이싱

t1 = (1, 2, "a", "b")
result = t1[:2]
print(result)
```

```python
# 튜플 더하기

t1 = (1, 2, "a", "b")
t2 = (3, 4)

print(t1 + t2)
print(t1)
```

```python
# 튜플 곱하기

t2 = (3, 4)
print(t2 * 3)
print(t2)
```
