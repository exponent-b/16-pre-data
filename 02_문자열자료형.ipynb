```python
# 문자열은 어떻게 만들고 사용할까?

## 1. 큰 따옴표로 양쪽 둘러싸기
a = "Hello"

## 2. 작은 따옴표로 양쪽 둘러싸기
b = 'Python is fun'

## 3. 큰 따옴표 3개로 둘러싸기
c = """Life is too short,"""

## 4. 작은 따옴표 3개로 둘러싸기
d = '''eee'''

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

```python
# 문자열 안에 작은 따옴표 포함하기

a = "python's favorite food is perl"
b = """python's favorite food is perl"""
c = '''python's favorite food is perl'''
d = 'python\'s favorite food is perl'
print(a)
print(b)
print(c)
print(d)
```

```python
# 문자열 안에 큰 따옴표 표현하기

a = '"pythom is very easy"he says.'
b = "\"pythom is very easy\"he says."

print(a)
print(b)
```

```python
mulitLine1 = """Life is too short, 
you need Python"""

mulitLine2 = """Life is too short, \n you need Python"""

print(mulitLine1)
print(mulitLine2)
```

```python
# 문자열 연산하기

a = "python "
b = "is fun"
print(a + b)
```

```python
a = "python"
print(a * 3)
```

```python
# 문자열 길이 구하기
a = "python"
print(len(a))
```

```python
# 문자열 인덱싱과 슬라이싱

a = "Life is too short, \n you need Python"
print(a[3])
print(len(a))
print(a[21])
print(a[-35])
print(a[0])
```

```python
# 문자열 슬라이싱

a = "Life is too short, you need Python"
print(a[:])
print(a[0:])
print(a[0:10])
print(a[:10])
print(a[0:10:2])
print(a[0:24:3])
print(a[12:17])
print(a[12:-5])
```

```python
# 문자열 포매팅
## 1. 숫자 바로 대입

print("I eat %d apples" %3)

## 2. 문자열 바로 대입
print("I eat %s apples" % "five")
```

```python
## 3. 숫자값을 나타내는 변수로 대입
num = 3
print("I eat %d apples" %num)
```

```python
# 2개 이상의 값 넣기

num = 10
day = "three"
print("I eat %d apples so I was sick for %s days" %(num, day))
```

```python
# format 함수를 활용한 포매팅

print("I eat {0} apples" .format(3))
print("I eat {0} apples" .format("five"))
```

```python
# format 함수를 이용해서 2개 이상의 값 넣기

num = 5
day = 10
print("I eat {0} apples so I was sick for {1} days".format(num, day))

'''
print("I eat {0}{1}{2} apples" .format(3, 2, 1))
print("I eat {0}{1}{0} apples" .format(3, 2, 1))
'''
```

```python
# f문자열 포매팅

name = "홍길동"
age = 10
print(f"나의 이름은 {name}입니다. 나이는 {age} 입니다.".format(name, age)) ##f안 넣으면 오류
```

```python
#문자열에 관련 함수들
a = "hobby"
print(type(a))

## 문자의 개수 세기 - count
print(a.count("b"))
```

```python
## 위치 알려주기 - find

a = "python is the best choice"
print(a.find("b"))
print(a.find("k")) # 없으면 -1
```

```python
# 위치 알려주기 2 - index

a = "Life is too short"
print(a.index("t"))
#print(a.index("k"))
```

```python
# 문자열 삽입 - join
print(",".join("abcd"))
```

```python
# 소문자를 대문자로 바꾸기 - upper
a = "abcd"
print(a.upper())
```

```python
# 대문자를 소문자로 바꾸기 - lower
a = "HI"
print(a.lower())
```

```python
# 공백 지우기 - 
a = "       hi          "
print(a.strip())
```

```python
# 공백 지우기 왼 / 오
print(a.lstrip())
print(a.rstrip())
```

```python
# 문자열 바꾸기 - replace

a = "life is too short"
print(a.replace("life", "your leg"))
```

```python
# 문자열 나누기 - split

a = "life is too short"
print(a.split())
```
