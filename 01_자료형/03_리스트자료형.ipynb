```python
# 리스트는  어떻게 만들고 사용할까?
# 리스트명 = [요소1, 요소2, 요소3, ...]

a = []
b = [1, 2, 3]
c = ["life", "is", "too", "short"]
d = [1, 2, "life", "is"]
e = [1, 2, ["life", "is"]]

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
```

```python
# 리스트의 인덱싱과 슬라이싱
a = [1, 2, 3]
print(a[0])
#print(a.index(1))
print(type(a[0]))
```

```python
print(a[0] + a[2])
```

```python
b = ["1", "2", "3", "a", "b", "c"]
print(b[-1])
b[-1]
```

```python
c = [1, 2, ["a", "b", ["life", "is"]]]
print(c[-1][-1][-2])
print(c[2][0])
print(len(c))
```

```python
# 리스트의 슬라이싱
a = [1, 2, 3, 4, 5]

print(a[:2]) # [시작:n-1:점프]
print(a[:3])
print(a[3:3])
print(a[2:])
```

```python
a = [1, 2, 3, ["a", "b", "c"], 4, 5]
# ["b", "c"]
print(len(a))
```

```python
a[3][1:2]
```

```python
# 리스트 더하기(+)
a = [1, 2, 3]
b = [4, 5, 6]

print(a + b)
```

```python
# 리스트 곱하기(*)
a = [1, 2, 3]
b = 3
print(a * b)
```

```python
# 리스트의 수정과 삭제
a = [1, 2, 3]
a[2] = 4
print(a)
```

```python
# del로 리스트 요소 삭제하기
a = [1, 2 ,3]
del a[1]
print(a)
```

```python
a = [1, 2, 3, 4, 5]
del a[2:]
print(a)
```

```python
# 리스트 관련 함수
## 리스트 요소 추가하기 -append

a = [1, 2, 3]
b = [5, 6]

print(a.append(b))
print(a)

a.append(b)
print(a)
```

```python
# 리스트 정렬 - sort
a = [1, 4, 2, 3]
a.sort()
print(a)

a.sort(reverse=True)
print(a)
a.sort(reverse=False)
print(a)
#error
re = a.sort()
print(re)
```

```python
result1 = sorted(a)
result2 = sorted(a, reverse=True)
result3 = sorted(a, reverse=False)

print(result1)
print(result2)
print(result3)
```

```python
# 리스트 뒤집기 - reverse
a = ["a","b", "c", "d"]
a.reverse()
print(a)
```

```python
# 인덱스 반환 - index
a = ["a","b", "c", "d"]
a.index("c")
```

```python
# 리스트에 요소 삽입 - insert
a = [1, 2, 3]
a.insert(0, 4)
print(a)
```

```python
# 리스트 요소 제거 - remove
a = [1, 2, 3, 1, 2, 3]

a.remove(3)
print(a)
```

```python
# 리스트 요소 꺼내기 - pop // 숫자 안 넣으면 가장 마지막 거
a = [1, 2, 3]
b = [1, 2, 3]
result1 = a.pop(1)
result2 = b.pop()
print(a)
print(b)
print(result1)
print(result2)
```

```python
# 리스트에 포함된 요소 x의 개수 세기 - count
a = [1, 2, 3, 1]
a.count(1)
```

```python
# 리스트 확장
a = [1, 2, 3]
a.extend([4, 5])
print(a)
```
