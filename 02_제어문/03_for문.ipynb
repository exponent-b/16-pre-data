```python
marks = [90, 25, 67, 45, 80]
num = 0
pas = 0
for score in marks:
	if score >= 60:
		num += 1
		pas += 1
		print(f"{num}번 학생 합격")
	else:
		num += 1
		print(f"{num}번 학생 불합격")
print(f"{pas}명 합격")
```

```python
# for문과 continue
marks = [90, 25, 67, 45, 80]
num = 0
for mark in marks:
	num += 1
	if mark < 60:
		continue
	
	print(f"{num}번 학생은 합격입니다.")
```

```python
# for문과 함께 자주 사용하는 range함수
a = range(10)
print(list(a))

b = range(1, 10)  ## ,n-1
print(list(b))

c = range(1, 10, 2)
print(list(c))

d = range(10, 1, -1)
print(list(d))
```

```python
# for와 range를 이용한 구구단
num = 1
for i in range(2, 9 + 1):
	num += 1
	print(f"============{num}단==============")
	for j in range(1, 9 + 1):
		print(f"\t{i} X {j} = {i * j}")
```

```python
# 리스트 컴프리헨션 사용하기
a = [1, 2, 3, 4]
result = []

for num in a :
	result.append(num * 3)

print(result)
```

```python
a = [1, 2, 3, 4]

result = [num * 3 for num in a]
print(result)
```

```python
# for문과 break문
for i in range(10):
	if i == 5:
		break
	print(i)

```

```python
# zip함수로 여러 리스트 순회하기
names = ['홍길동', '김철수', '이영희']
scores = [85, 92, 78]

for name, score in zip(names, scores):
	print(f"{name}: {score}점")
```
