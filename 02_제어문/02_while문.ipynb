```python
# while문의 기본구조
#while 조건문:
#	수행할 문장1
#	수행할 문장2
```

```python
treehit = 0
while treehit < 10:
	treehit = treehit + 1
	print(f"나무를 {treehit}번 찍었습니다")
	if treehit == 10:
		print("나무가 넘어집니다")
```

```python
coffee = 5
money = 300
while money:
	print("돈을 받았으니 커피를 줍니다")
	coffee -= 1
	print(f"남은 커피는{coffee}개 입니다.")
	

	if coffee == 0:
		print("커피가 다 떨어졌습니다.")
		break
```

```python
coffee = 5
while True:
	money = int(input("돈을 넣어주세요: "))
	if money == 300:
		print("커피 줄게")
		coffee -= 1
	elif money > 300:
		print(f"거스름돈 {money - 300}")
		coffee -= 1
	else:
		print("돈 반환, 커피 안 줌")
		print(f"현재 남은 커피양은 {coffee}개 입니다.")

	if coffee == 0:
		print("커피 없다")
		break
```

```python
# while문의 맨 처음으로 돌아가기
a = 0
while a < 10:
	a += 1
	if a % 2 == 0:
		continue
	print(a)
```
