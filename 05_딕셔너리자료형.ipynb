```python
# 딕셔너리는 어떻게 만들까?
##{key1:value1, key2:vlaue2, ...}

dic = {'name':"pey", "phone":"010-1234-1234", "birth":"1118"}

print(dic)
print(type(dic))
```

```python
# 딕셔너리의 키 값으로 올 수 없는 데이터 타입은 리스트 -> mutable
a = [1, 2, 3]
a[0] = 3
print(a)

a = 1
a = 2
print(a)
```

```python
# 딕셔너리 쌍 추가하기
a = {1:"a"}
a[2] = "b"
print(a)
```

```python
a[3] = [1, 2, 3]
print(a)
```

```python
# 딕셔너리 요소 삭제하기
print(a)
del a[1]
print(a)
```

```python
# 딕셔너리에 key를 사용하여 vlaue 얻기
grade = {"pey" : 10, "juli" : 99}

print(grade["pey"])
print(grade["juli"])
```

```python
# 딕셔너리 관련 함수
## 1. key 리스트 만들기 - keys
dic = {'name':"pey", "phone":"010-1234-1234", "birth":"1118"}
print(dic.keys())
```

```python
# values 리스트 만들기 - values
print(list(dic.values()))
print(dic.values())
```

```python
# key와 value의 쌍 얻기 - items
print(dic)
print(dic.items())
print(list(dic.items()))
```

```python
# key와 value의 쌍을 모두 지우기 - clear
a.clear()
print(a)
```

```python
# key로 vlaue 얻기 - get
dic = {'name':"pey", "phone":"010-1234-1234", "birth":"1118"}
print(dic["name"])
print(dic.get("name"))
print(dic)
```

```python
#print(a["hobby"]) # error
'''@@@@@@@@@'''
print(a.get("hobby", 1)) #-> none이면 1반환
```

```python
# 해당 key가 딕셔너리 안에 있는지 조사하기 - in
print(dic)
print("name" in dic)
print("email" in dic)
```

```python
# key로 value 꺼내기 - pop
dic = {'name':"pey", "phone":"010-1234-1234", "birth":"1118"}
re = dic.pop("phone")
print(re)
print(dic)
```
